## MERGE FOLDER

```python
import os
from PIL import Image

# Create a new directory to store the resulting images
output_folder = '/content/new/images'
os.makedirs(output_folder, exist_ok=True)
```

```python
import os
import shutil

def merge_folders(source_folders, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for folder in source_folders:
        # Walk through each file and subdirectory in the source folder
        for root, dirs, files in os.walk(folder):
            for file in files:
                src_path = os.path.join(root, file)
                dest_path = os.path.join(destination_folder, file)

                # Copy the file to the destination folder
                shutil.copy(src_path, dest_path)

source_folder1 = "/content/images/train"
source_folder2 = '/content/images/test'
destination_folder = '/content/new/images'

# Merge contents of source_folder1 and source_folder2 into destination_folder
merge_folders([source_folder1, source_folder2], destination_folder)
```

```python
import zipfile
import os

def zip_folder(folder_path, output_path):
    with zipfile.ZipFile(output_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                zipf.write(file_path, os.path.relpath(file_path, folder_path))
```

```python
# Example usage
folder_to_zip = "/content/new"
output_zip_path = '/content/one-slide_1.zip'

zip_folder(folder_to_zip, output_zip_path)
```
